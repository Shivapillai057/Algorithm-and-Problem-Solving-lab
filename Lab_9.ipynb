# Shortest Path Using Dijkstra Algorithm in Python
Problem Description:
Given a weighted graph with non negative edge weights. Find the shortest distance from a source vertex to all other vertices using Dijkstra algorithm

#Theory
Dijkstra algorithm finds shortest distance from one source node to all other nodes in a weighted graph with non negative weights. Dictionary structure stores nodes as keys and neighbor nodes with weights as values.

##Algorithm Steps

1. Represent graph using dictionary.
2. Set distance of every node as infinity.
3. Set distance of source node as 0.
4. Select node with smallest distance among unvisited nodes.
5. Update distance of neighbor nodes.
6. Repeat until all nodes get visited.

```python
def dijkstra(graph, start):
    dist = {node: float('inf') for node in graph}
    prev = {}
    visited = []

    dist[start] = 0

    while len(visited) < len(graph):
        min_node = None
        for node in graph:
            if node not in visited and (min_node is None or dist[node] < dist[min_node]):
                min_node = node

        visited.append(min_node)

        for neighbor in graph[min_node]:
            new_dist = dist[min_node] + graph[min_node][neighbor]
            if new_dist < dist[neighbor]:
                dist[neighbor] = new_dist
                prev[neighbor] = min_node

    return dist, prev


def path(prev, start, end):
    p = []
    while end != start:
        p.append(end)
        end = prev.get(end)
    p.append(start)
    return p[::-1]

```

```python
def dijkstra(graph, start):
    dist = {node: float('inf') for node in graph}
    prev = {node: None for node in graph}
    dist[start] = 0
    visited = set()

    while len(visited) < len(graph):
        current = None
        current_dist = float('inf')

        for node in graph:
            if node not in visited and dist[node] < current_dist:
                current = node
                current_dist = dist[node]

        if current is None:
            break

        visited.add(current)

        for neighbor, weight in graph[current].items():
            new_dist = dist[current] + weight
            if new_dist < dist[neighbor]:
                dist[neighbor] = new_dist
                prev[neighbor] = current

    return dist, prev


def get_path(prev, start, end):
    path = []
    while end is not None:
        path.append(end)
        end = prev[end]
    path.reverse()
    if path[0] == start:
        return path
    return []


graph = {}
n = int(input("Enter number of nodes: "))

for i in range(n):
    node = input(f"Enter node {i+1}: ")
    graph[node] = {}
    e = int(input(f"Enter number of edges for {node}: "))
    for j in range(e):
        nb = input("Neighbor: ")
        w = int(input("Weight: "))
        graph[node][nb] = w


extra = int(input("Enter number of extra nodes: "))
for i in range(extra):
    node = input(f"Extra node {i+1}: ")
    graph[node] = {}
    e = int(input(f"Edges for {node}: "))
    for j in range(e):
        nb = input("Neighbor: ")
        w = int(input("Weight: "))
        graph[node][nb] = w


while True:
    start = input("\nEnter source node (exit to stop): ")
    if start == "exit":
        break

    if start not in graph:
        print("Node not found")
        continue

    dist, prev = dijkstra(graph, start)

    for node in graph:
        path = get_path(prev, start, node)
        print(f"{start} -> {node} = {dist[node]} | Path: {' -> '.join(path)}")
```
